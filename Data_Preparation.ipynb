```python
import pandas as pd
import numpy as np
```

```python
df_CS = pd.read_csv('/content/drive/MyDrive/Shiller Case-US/CSUSHPISA.csv')

df_CS["DATE"] = pd.to_datetime(df_CS["DATE"])
limit = df_CS["DATE"] < "2024-01-01"
df_CS = df_CS[limit]

df_CS.reset_index(inplace = True)
df_CS.drop(columns = ["index"], inplace = True)

# Creating "Year" and "Month" columns
df_CS["Year"] = pd.DatetimeIndex(df_CS["DATE"]).year
df_CS["Month"] = pd.DatetimeIndex(df_CS["DATE"]).month
print("Shape of the CASE-SHILLER Index:- ", df_CS.shape)
df_CS
```

```python
# Unemployment DataSet

df_unemp = pd.read_csv('/content/drive/MyDrive/Shiller Case-US/UNRATE.csv')

df_unemp["DATE"] = pd.to_datetime(df_unemp["DATE"])
limit = df_unemp["DATE"] < "2024-01-01"
df_unemp = df_unemp[limit]
limit = df_unemp["DATE"] >= "1987-01-01"
df_unemp = df_unemp[limit]

df_unemp
```

```python
#Employment DataSet
df_emp = pd.read_csv('/content/drive/MyDrive/Shiller Case-US/Employment.csv')

df_emp["DATE"] = pd.to_datetime(df_emp["DATE"])
df_emp = df_emp.rename(columns={'LREM64TTUSM156S':'EmployRate'})
limit = df_emp["DATE"] < "2024-01-01"
df_emp = df_emp[limit]
limit = df_emp["DATE"] >= "1987-01-01"
df_emp = df_emp[limit]
df_emp = df_emp.reset_index(drop=True)

df_emp
```

```python
df_pcgdp = pd.read_csv('/content/drive/MyDrive/Shiller Case-US/A939RX0Q048SBEA.csv')

df_pcgdp = df_pcgdp.rename(columns={'A939RX0Q048SBEA': 'Per_Capita_GDP'})
df_pcgdp["DATE"] = pd.to_datetime(df_pcgdp["DATE"])
limit = df_pcgdp["DATE"] < "2024-01-01"
df_pcgdp = df_pcgdp[limit]
limit = df_pcgdp["DATE"] >= "1987-01-01"
df_pcgdp = df_pcgdp[limit]

df_pcgdp
```

```python
#Fderal Funds DataSet
df_fed = pd.read_csv('/content/drive/MyDrive/Shiller Case-US/FEDFUNDS.csv')

df_fed["DATE"] = pd.to_datetime(df_fed["DATE"])
limit = df_fed["DATE"] < "2024-01-01"
df_fed = df_fed[limit]
limit = df_fed["DATE"] >= "1987-01-01"
df_fed = df_fed[limit]
df_fed = df_fed.reset_index(drop=True)

df_fed
```

```python
#Construction Price DataSet

df_const = pd.read_csv('/content/drive/MyDrive/Shiller Case-US/WPUSI012011.csv')

df_const = df_const.rename(columns={'WPUSI012011': 'Construction_Price'})
df_const["DATE"] = pd.to_datetime(df_const["DATE"])
limit = df_const["DATE"] < "2024-01-01"
df_const = df_const[limit]
limit = df_const["DATE"] >= "1987-01-01"
df_const = df_const[limit]
df_const = df_const.reset_index(drop=True)

df_const
```

```python
#Interest Rate

df_interest = pd.read_csv('/content/drive/MyDrive/Shiller Case-US/INTDSRUSM193N.csv')

df_interest = df_interest.rename(columns={'INTDSRUSM193N': 'Interest_Rate'})
df_interest["DATE"] = pd.to_datetime(df_interest["DATE"])
limit = df_interest["DATE"] < "2024-01-01"
df_interest = df_interest[limit]
limit = df_interest["DATE"] >= "1987-01-01"
df_interest = df_interest[limit]
df_interest = df_interest.reset_index(drop=True)

df_interest
```

```python
# Total household

df_totalHouse = pd.read_csv('/content/drive/MyDrive/Shiller Case-US/TTLHH.csv')

df_totalHouse["DATE"] = pd.to_datetime(df_totalHouse["DATE"])
limit = df_totalHouse["DATE"] < "2024-01-01"
df_totalHouse = df_totalHouse[limit]
limit = df_totalHouse["DATE"] >= "1987-01-01"
df_totalHouse = df_totalHouse[limit]
df_totalHouse = df_totalHouse.reset_index(drop=True)

df_totalHouse.tail()
```

```python
#Subsidy

df_subsidy = pd.read_csv('/content/drive/MyDrive/Shiller Case-US/L312051A027NBEA.csv')

df_subsidy = df_subsidy.rename(columns={'L312051A027NBEA': 'Subsidy'})
df_subsidy["DATE"] = pd.to_datetime(df_subsidy["DATE"])
limit = df_subsidy["DATE"] < "2024-01-01"
df_subsidy = df_subsidy[limit]
limit = df_subsidy["DATE"] >= "1987-01-01"
df_subsidy = df_subsidy[limit]
df_subsidy = df_subsidy.reset_index(drop=True)

print(df_subsidy.shape)
df_subsidy.tail()
```

```python
# GDP

df_gdp = pd.read_csv('/content/drive/MyDrive/Shiller Case-US/GDP.csv')

df_gdp["DATE"] = pd.to_datetime(df_gdp["DATE"])
limit = df_gdp["DATE"] < "2024-01-01"
df_gdp = df_gdp[limit]
limit = df_gdp["DATE"] >= "1987-01-01"
df_gdp = df_gdp[limit]
df_gdp = df_gdp.reset_index(drop=True)

df_gdp
```

```python
# Labour

df_labour = pd.read_csv('/content/drive/MyDrive/Shiller Case-US/LFWA64TTUSM647S.csv')

df_labour.rename(columns={'LFWA64TTUSM647S':'Labour_stats'}, inplace = True)
df_labour["DATE"] = pd.to_datetime(df_labour["DATE"])
limit = df_labour["DATE"] < "2024-01-01"
df_labour = df_labour[limit]
limit = df_labour["DATE"] >= "1987-01-01"
df_labour = df_labour[limit]
df_labour = df_labour.reset_index(drop=True)

df_labour
```

```python
df_CS = pd.merge(df_CS,df_pcgdp, how = "left")
df_CS = pd.merge(df_CS, df_interest, how="left")
df_CS.head()
```

```python
# Concating dataframes having monthly data to create one dataframe

df = pd.DataFrame()
df_bymonth = [df_CS, df_unemp, df_const, df_emp, df_labour, df_fed]
for df1 in df_bymonth:
    df1 = df1.set_index("DATE")
    df = pd.concat([df,df1], axis = 1)
print(df.shape)
df
```

```python
df.isna().sum()
```

```python
df["Per_Capita_GDP"] = df["Per_Capita_GDP"].interpolate()
df = df.dropna()
df.isna().sum()
```
